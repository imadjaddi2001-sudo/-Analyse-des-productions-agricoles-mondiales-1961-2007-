## ÉTAPE 1 — Charger et explorer le dataset
Dataset : Productions agricoles mondiales (FAO)


```python
import pandas as pd  # La bibliothèque principale pour manipuler des données
```

## 1. CHARGER LE FICHIER CSV


```python
# pd.read_csv() lit le fichier et le transforme en DataFrame (tableau)
df = pd.read_csv("JEU_DE_DONNEE.csv")

print("=== Fichier chargé avec succès ! ===\n")
```

## 2. VOIR LES PREMIÈRES LIGNES


```python
# .head(10) affiche les 10 premières lignes
# Utile pour voir à quoi ressemblent les données
print("--- Les 10 premières lignes ---")
print(df.head(10))
print()
```

## 3. DIMENSIONS DU TABLEAU


```python
# .shape retourne (nombre de lignes, nombre de colonnes)
lignes, colonnes = df.shape
print(f"--- Dimensions du tableau ---")
print(f"Nombre de lignes   : {lignes:,}")
print(f"Nombre de colonnes : {colonnes}")
print()
```

## 4. NOMS ET TYPES DES COLONNES


```python
# .dtypes indique le type de chaque colonne
# object = texte, float64 = nombre décimal, int64 = entier
print("--- Colonnes et leurs types ---")
print(df.dtypes)
print()
```

## 5. VALEURS MANQUANTES


```python
# .isnull() détecte les cases vides
# .sum() les compte colonne par colonne
print("--- Valeurs manquantes par colonne ---")
print(df.isnull().sum())
print()
```

## 6. STATISTIQUES RAPIDES


```python
# .describe() calcule automatiquement : count, mean, min, max, etc.
# Uniquement sur les colonnes numériques
print("--- Statistiques sur la colonne 'value' ---")
print(df["value"].describe())
print()
```

## 7. VALEURS UNIQUES DANS CHAQUE COLONNE CLÉ


```python
# .nunique() compte le nombre de valeurs différentes
print("--- Combien de valeurs uniques ? ---")
print(f"Pays / zones    : {df['country_or_area'].nunique()}")
print(f"Cultures        : {df['category'].nunique()}")
print(f"Types de mesure : {df['element'].nunique()}")
print(f"Années          : {df['year'].nunique()}")
print()
```

## 8. APERÇU DES VALEURS UNIQUES


```python
# .unique() liste toutes les valeurs différentes d'une colonne
print("--- Types de mesures disponibles (element) ---")
print(df["element"].unique())
print()

print("--- Exemples de cultures (category) ---")
print(df["category"].unique()[:20])  # Seulement les 20 premières
print()

print("--- Plage d'années ---")
print(f"De {df['year'].min()} à {df['year'].max()}")
```
