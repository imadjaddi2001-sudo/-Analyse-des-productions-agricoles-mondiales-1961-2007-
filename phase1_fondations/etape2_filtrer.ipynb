## ÉTAPE 2 — Filtrer et sélectionner des données
Objectif : blé (wheat) + données mondiales (World +)


```python
import pandas as pd
```

## 1. CHARGER LE FICHIER


```python
df = pd.read_csv("JEU_DE_DONNEE.csv")
print(f"Dataset chargé : {df.shape[0]:,} lignes au total\n")
```

## 2. EXPLORER LES VALEURS UNIQUES AVANT DE FILTRER


## Bonne habitude : toujours vérifier les valeurs exactes
avant de filtrer (orthographe, majuscules, espaces...)


```python
print("--- Valeurs uniques dans 'category' contenant 'wheat' ---")
wheat_values = df[df["category"].str.contains("wheat", case=False, na=False)]["category"].unique()
print(wheat_values)
print()

print("--- Valeurs uniques dans 'country_or_area' contenant 'World' ---")
world_values = df[df["country_or_area"].str.contains("World", case=False, na=False)]["country_or_area"].unique()
print(world_values)
print()
```

## 3. FILTRER SUR LE BLÉ (wheat)


```python
# On garde uniquement les lignes où category == "Wheat"
# Le == fait une comparaison exacte
filtre_ble = df["category"] == "wheat"
df_ble = df[filtre_ble]

print(f"--- Après filtre blé ---")
print(f"Lignes restantes : {df_ble.shape[0]:,}")
print(df_ble.head(5))
print()
```

## 4. FILTRER SUR LE MONDE (World +)


```python
# On garde uniquement la ligne "World +"
filtre_monde = df["country_or_area"] == "World +"
df_monde = df[filtre_monde]

print(f"--- Après filtre mondial ---")
print(f"Lignes restantes : {df_monde.shape[0]:,}")
print(df_monde.head(5))
print()
```

## 5. COMBINER LES DEUX FILTRES (blé ET mondial)


```python
# L'opérateur & combine deux conditions (ET logique)
# Parenthèses obligatoires autour de chaque condition !
df_ble_monde = df[
    (df["category"] == "wheat") &
    (df["country_or_area"] == "World +")
]

print(f"--- Blé mondial ---")
print(f"Lignes restantes : {df_ble_monde.shape[0]:,}")
print(df_ble_monde)
print()
```

## 6. FILTRER UN TYPE DE MESURE PRÉCIS


```python
# 3 types existent : "Area Harvested", "Yield", "Production Quantity"
# On se concentre sur la quantité produite

df_production = df[
    (df["category"] == "wheat") &
    (df["country_or_area"] == "World +") &
    (df["element"] == "Production Quantity")
]

print(f"--- Production de blé mondiale uniquement ---")
print(df_production[["year", "value", "unit"]].sort_values("year"))
print()
```

## 7. UTILISER query() — syntaxe alternative plus lisible


```python
# query() permet d'écrire les conditions comme une phrase
df_query = df.query(
    "category == 'wheat' and country_or_area == 'World +' and element == 'Production Quantity'"
)

print("--- Même résultat avec query() ---")
print(df_query[["year", "value", "unit"]].sort_values("year").head(10))
print()
```

## 8. SAUVEGARDER LE RÉSULTAT FILTRÉ


```python
# On exporte le tableau filtré dans un nouveau fichier CSV
df_production.to_csv("ble_mondial_production.csv", index=False)
print("Fichier 'ble_mondial_production.csv' sauvegardé ✓")
print(f"Il contient {len(df_production)} lignes (une par année)")
```
