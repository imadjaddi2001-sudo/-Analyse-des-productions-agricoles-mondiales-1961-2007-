## ÉTAPE 4 — Visualisation avec matplotlib
Objectif : évolution de la production mondiale de blé 1961-2007


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## 1. CHARGER ET FILTRER LES DONNÉES


```python
df = pd.read_csv("JEU_DE_DONNEE.csv")

df_ble = df[
    (df["category"] == "wheat") &
    (df["country_or_area"] == "World +") &
    (df["element"] == "Production Quantity")
].dropna(subset=["value"]).sort_values("year")

# Convertir en millions de tonnes pour plus de lisibilité
df_ble["valeur_millions"] = df_ble["value"] / 1_000_000

annees = df_ble["year"]
production = df_ble["valeur_millions"]

print(f"Données prêtes : {len(df_ble)} années (1961 → 2007)\n")
```

## 2. GRAPHIQUE SIMPLE — La base


```python
plt.figure(figsize=(12, 5))

plt.plot(annees, production)               # tracer la courbe
plt.title("Production mondiale de blé")   # titre
plt.xlabel("Année")                        # axe X
plt.ylabel("Production (millions tonnes)") # axe Y
plt.show()
```

## 3. GRAPHIQUE AMÉLIORÉ — Version propre


```python
fig, ax = plt.subplots(figsize=(13, 6))

# --- Courbe principale ---
ax.plot(
    annees, production,
    color="#2563EB",       # bleu
    linewidth=2.5,
    marker="o",            # point à chaque année
    markersize=4,
    label="Production (Mt)"
)

# --- Zone colorée sous la courbe ---
ax.fill_between(annees, production, alpha=0.1, color="#2563EB")

# --- Annoter le min et le max ---
idx_max = production.idxmax()
idx_min = production.idxmin()

ax.annotate(
    f"Max : {production[idx_max]:.0f} Mt\n({int(annees[idx_max])})",
    xy=(annees[idx_max], production[idx_max]),
    xytext=(annees[idx_max] - 8, production[idx_max] - 30),
    arrowprops=dict(arrowstyle="->", color="green"),
    color="green", fontsize=10, fontweight="bold"
)

ax.annotate(
    f"Min : {production[idx_min]:.0f} Mt\n({int(annees[idx_min])})",
    xy=(annees[idx_min], production[idx_min]),
    xytext=(annees[idx_min] + 3, production[idx_min] + 40),
    arrowprops=dict(arrowstyle="->", color="red"),
    color="red", fontsize=10, fontweight="bold"
)

# --- Ligne de tendance moyenne ---
moyenne = production.mean()
ax.axhline(
    y=moyenne,
    color="orange", linestyle="--", linewidth=1.5,
    label=f"Moyenne : {moyenne:.0f} Mt"
)

# --- Mise en forme ---
ax.set_title(
    "Production mondiale de blé (1961 – 2007)",
    fontsize=16, fontweight="bold", pad=15
)
ax.set_xlabel("Année", fontsize=12)
ax.set_ylabel("Production (millions de tonnes)", fontsize=12)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:.0f} Mt"))
ax.set_xlim(1961, 2007)
ax.grid(axis="y", linestyle="--", alpha=0.4)
ax.legend(fontsize=11)

plt.tight_layout()

# --- Sauvegarder le graphique ---
plt.savefig("graphique_ble_mondial.png", dpi=150, bbox_inches="tight")
print("Graphique sauvegardé : 'graphique_ble_mondial.png' ✓")

plt.show()
```

## 4. QUELQUES STATISTIQUES EN BONUS


```python
print("\n--- Statistiques clés ---")
print(f"Production minimale : {production.min():.0f} Mt en {int(annees[idx_min])}")
print(f"Production maximale : {production.max():.0f} Mt en {int(annees[idx_max])}")
print(f"Moyenne sur la période : {moyenne:.0f} Mt")

# Croissance entre 1961 et 2007
debut = production.iloc[0]
fin   = production.iloc[-1]
croissance = ((fin - debut) / debut) * 100
print(f"Croissance 1961 → 2007 : +{croissance:.0f}%")
```
